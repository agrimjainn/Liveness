```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    success, frame = cap.read()

    cv2.putText(
    frame,
    "Hello Camera",
    (50,50),
    cv2.FONT_HERSHEY_COMPLEX,
    1,
    (255,255,255),
    2
);
    
    cv2.imshow("Camera", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()


# ho
# from telegram
```

```python
cap = cv2.VideoCapture(0)
count = 0

eye_open = True
prev_eye_open = True

while True:
    success, frame = cap.read()

    cv2.imshow("Camera", frame)

    key = cv2.waitKey(1) & 0xFF

    if key == ord('o'):
        eye_open = True
    if key == ord('c'):
        eye_open = False

    if not prev_eye_open and eye_open:
        count += 1
        print(count)

    prev_eye_open = eye_open

    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()


```
